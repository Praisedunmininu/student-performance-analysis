## Objective
Analyze student performance to understand which factors affect grades.

```python

# 2. Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

# 3. Load Dataset
df = pd.read_csv("student_performance.csv")  # Make sure your CSV path is correct
df.head()   # Show first 5 rows
df.info()   # Overview of dataset
```

 Markdown (Data Cleaning / Preprocessing)
## 4. Data Cleaning / Preprocessing
- Handle missing values and duplicates
- Check column types
- Create new columns if needed


```python

# Create categorical score based on total_score
bins = [0, 60, 84, 100]
labels = ["low", "average", "high"]
df["student_score"] = pd.cut(df["total_score"], bins=bins, labels=labels)
df.head()
```

## Exploratory Data Analysis (EDA)
- Descriptive statistics: df.describe()
- Grouping by grade or other columns
- Counts, value_counts, averages

```python
# (Descriptive Stats & Grouping0
# Average total score by grade
df.groupby("grade")["total_score"].mean()
```

 
## Visualization

```python
  # Scatter Plot Example

plt.scatter(df["weekly_self_study_hours"], df["total_score"])
plt.title("Weekly Self-Study Hours vs Total Score")
plt.xlabel("Weekly Self-Study Hours")
plt.ylabel("Total Score")
plt.show()
```


##  Correlation Analysis
Check correlations between variables to understand relationships.
- Weekly self-study hours have the strongest positive correlation with total score.
- Attendance and class participation have minimal correlation.

```python
# Correlation
df[["weekly_self_study_hours","attendance_percentage","class_participation","total_score"]].corr()
```
